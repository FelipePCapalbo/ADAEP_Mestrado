```python
# Addition (+)
num1 = 10
num2 = 5
result_add = num1 + num2
print(f"Addition: {num1} + {num2} = {result_add}")

# Subtraction (-)
num3 = 15
num4 = 7
result_sub = num3 - num4
print(f"Subtraction: {num3} - {num4} = {result_sub}")

# Multiplication (*)
num5 = 4
num6 = 6
result_mul = num5 * num6
print(f"Multiplication: {num5} * {num6} = {result_mul}")

# Division (/)
num7 = 20
num8 = 4
result_div = num7 / num8
print(f"Division: {num7} / {num8} = {result_div}")

# Floor Division (//) - Returns the quotient (integer part) of the division
num9 = 23
num10 = 5
result_floor_div = num9 // num10
print(f"Floor Division: {num9} // {num10} = {result_floor_div}")

# Modulo (%) - Returns the remainder of the division
num11 = 17
num12 = 3
result_modulo = num11 % num12
print(f"Modulo: {num11} % {num12} = {result_modulo}")

# Exponentiation (**)
base = 2
exponent = 3
result_exp = base ** exponent
print(f"Exponentiation: {base} ** {exponent} = {result_exp}")
```

```python
# prompt: Faca um script utilizando operadores para strings. Utilize as palavras em portugues

# Concatenação (+)
palavra1 = "Olá"
palavra2 = "Mundo"
resultado_concatenacao = palavra1 + " " + palavra2
print(f"Concatenação: {palavra1} + {palavra2} = {resultado_concatenacao}")

# Repetição (*)
palavra3 = "Python"
repeticoes = 3
resultado_repeticao = palavra3 * repeticoes
print(f"Repetição: {palavra3} * {repeticoes} = {resultado_repeticao}")

# Verificação de Substring (in)
frase = "O rato roeu a roupa do rei de Roma"
substring = "rei"
resultado_substring = substring in frase
print(f"'{substring}' está em '{frase}': {resultado_substring}")

# Verificação de Substring (not in)
frase = "O rato roeu a roupa do rei de Roma"
substring = "rainha"
resultado_substring = substring not in frase
print(f"'{substring}' não está em '{frase}': {resultado_substring}")
```

**Métodos de manipulação de Strings**


```python
# 1. upper() e lower() - Alterando o Caso:
texto = "Olá, Mundo!"
texto_maiusculo = texto.upper()
texto_minusculo = texto.lower()
print(f"Maiúsculo: {texto_maiusculo}")
print(f"Minúsculo: {texto_minusculo}")

# 2. strip() - Removendo Espaços em Branco:
texto_com_espacos = "   Esta string tem espaços.   "
texto_sem_espacos = texto_com_espacos.strip()
print(f"Texto sem espaços: '{texto_sem_espacos}'")

# 3. split() - Dividindo em uma Lista:
frase = "Esta é uma frase."
palavras = frase.split()
print(f"Palavras: {palavras}")

# 4. join() - Unindo Elementos de uma Lista:
palavras = ["Esta", "é", "uma", "lista"]
string_unida = " ".join(palavras)
print(f"String unida: {string_unida}")

# 5. replace() - Substituindo Substrings:
texto = "Este é um texto com palavras antigas."
novo_texto = texto.replace("antigas", "novas")
print(f"Novo texto: {novo_texto}")

# 6. find() - Encontrando o Índice de uma Substring:
texto = "Esta é uma frase."
indice = texto.find("frase")
print(f"Índice de 'frase': {indice}")

# 7. startswith() e endswith() - Verificando Prefixos e Sufixos:
texto = "Olá, mundo!"
comeca_com_ola = texto.startswith("Olá")
termina_com_exclamacao = texto.endswith("!")
print(f"Começa com 'Olá': {comeca_com_ola}")
print(f"Termina com '!': {termina_com_exclamacao}")

# 8. isdigit(), isalpha(), isalnum() - Verificações de Tipo de Caractere:
texto1 = "12345"
texto2 = "Olá"
texto3 = "Olá123"
print(f"'{texto1}' é todo dígitos: {texto1.isdigit()}")
print(f"'{texto2}' é todo alfabético: {texto2.isalpha()}")
print(f"'{texto3}' é alfanumérico: {texto3.isalnum()}")
```

Tuplas são ordenadas e imutáveis durante a execução


```python
# prompt: Faça uma exemplo utilizando tupla, conjuto e lista e dicionário com um e dois níveis. Adicione um comentario explicando as caracteristicas de cada estrutura

# Tupla: ordenada e imutável
tupla_frutas = ("maçã", "banana", "laranja")
print(tupla_frutas)

# Conjunto: não ordenado e não permite elementos duplicados
conjunto_cores = {"vermelho", "verde", "azul", "vermelho"}
print(conjunto_cores)

# Lista: ordenada e mutável
lista_numeros = [1, 2, 3, 4, 5]
print(lista_numeros)

# Dicionário de um nível: pares chave-valor
dicionario_pessoa = {"nome": "Maria", "idade": 30, "cidade": "São Paulo"}
print(dicionario_pessoa)

# Dicionário de dois níveis:
dicionario_pessoas = {
  "pessoa1": {"nome": "Maria", "idade": 30},
  "pessoa2": {"nome": "João", "idade": 25}
}
print(dicionario_pessoas)
```

```python
# prompt: Dê um exemplo das operações que podem ser feitas com tuplas

# Acessando elementos de uma tupla
tupla_frutas = ("maçã", "banana", "laranja")
print(tupla_frutas[0])  # Saída: maçã

# Fatiamento de tupla
print(tupla_frutas[1:3])  # Saída: ('banana', 'laranja')

# Concatenando tuplas
tupla1 = (1, 2, 3)
tupla2 = (4, 5, 6)
tupla_concatenada = tupla1 + tupla2
print(tupla_concatenada)  # Saída: (1, 2, 3, 4, 5, 6)

# Verificando se um elemento existe na tupla
if "maçã" in tupla_frutas:
  print("Maçã está na tupla")

# Iterando sobre uma tupla
for fruta in tupla_frutas:
  print(fruta)
```

```python
# prompt: Dê um exemplo das operações que podem ser feitas com conjuntos

# União de conjuntos
conjunto1 = {1, 2, 3}
conjunto2 = {3, 4, 5}
conjunto_uniao = conjunto1 | conjunto2
print(conjunto_uniao)  # Saída: {1, 2, 3, 4, 5}

# Interseção de conjuntos
conjunto_intersecao = conjunto1 & conjunto2
print(conjunto_intersecao)  # Saída: {3}

# Diferença de conjuntos
conjunto_diferenca = conjunto1 - conjunto2
print(conjunto_diferenca)  # Saída: {1, 2}

# Adição e remoção de elementos
conjunto1.add(4)
print(conjunto1)  # Saída: {1, 2, 3, 4}

conjunto1.remove(2)
print(conjunto1)  # Saída: {1, 3, 4}
```

```python
# prompt: Dê um exemplo das operações que podem ser feitas com listas incluindo o opcao de adicionar outra lista na lista

# Criando uma lista
numeros = [1, 2, 3, 4, 5]

# Acessando elementos
print(numeros[0])  # Saída: 1
print(numeros[-1]) # Saída: 5

# Fatiamento
print(numeros[1:3])  # Saída: [2, 3]

# Modificando elementos
numeros[2] = 10
print(numeros)  # Saída: [1, 2, 10, 4, 5]

# Adicionando elementos
numeros.append(6)
print(numeros)  # Saída: [1, 2, 10, 4, 5, 6]

# Inserindo elementos em uma posição específica
numeros.insert(1, 7)
print(numeros)  # Saída: [1, 7, 2, 10, 4, 5, 6]

# Removendo elementos
numeros.remove(10)  # Remove o elemento com valor 10
print(numeros)  # Saída: [1, 7, 2, 4, 5, 6]

# Removendo elemento por índice
del numeros[0]
print(numeros)  # Saída: [7, 2, 4, 5, 6]

# Ordenando a lista
numeros.sort()
print(numeros)  # Saída: [2, 4, 5, 6, 7]

# Invertendo a ordem da lista
numeros.reverse()
print(numeros)  # Saída: [7, 6, 5, 4, 2]

# Tamanho da lista
print(len(numeros))  # Saída: 5

# Adicionando outra lista na lista
outra_lista = [8, 9, 10]
numeros.extend(outra_lista)
print(numeros)  # Saída: [7, 6, 5, 4, 2, 8, 9, 10]
```

```python
# prompt: Dê um exemplo das operações que podem ser feitas com dicionários com mais de um nível. Apresente em comentários as saídas

# Dicionário de dois níveis:
dicionario_pessoas = {
  "pessoa1": {"nome": "Maria", "idade": 30},
  "pessoa2": {"nome": "João", "idade": 25}
}

# Acessando valores
print(dicionario_pessoas["pessoa1"]["nome"])  # Saída: Maria
print(dicionario_pessoas["pessoa2"]["idade"])  # Saída: 25

# Modificando valores
dicionario_pessoas["pessoa1"]["idade"] = 31
print(dicionario_pessoas)
# Saída: {'pessoa1': {'nome': 'Maria', 'idade': 31}, 'pessoa2': {'nome': 'João', 'idade': 25}}

# Adicionando um novo par chave-valor
dicionario_pessoas["pessoa1"]["cidade"] = "São Paulo"
print(dicionario_pessoas)
# Saída: {'pessoa1': {'nome': 'Maria', 'idade': 31, 'cidade': 'São Paulo'}, 'pessoa2': {'nome': 'João', 'idade': 25}}

# Removendo um par chave-valor
del dicionario_pessoas["pessoa2"]["idade"]
print(dicionario_pessoas)
# Saída: {'pessoa1': {'nome': 'Maria', 'idade': 31, 'cidade': 'São Paulo'}, 'pessoa2': {'nome': 'João'}}

# Iterando sobre as chaves do dicionário externo
for chave in dicionario_pessoas:
  print(chave)
  # Saída: pessoa1
  # Saída: pessoa2

# Iterando sobre os pares chave-valor do dicionário externo
for chave, valor in dicionario_pessoas.items():
  print(chave, valor)
  # Saída: pessoa1 {'nome': 'Maria', 'idade': 31, 'cidade': 'São Paulo'}
  # Saída: pessoa2 {'nome': 'João'}

# Iterando sobre os pares chave-valor do dicionário interno
for chave, valor in dicionario_pessoas["pessoa1"].items():
  print(chave, valor)
  # Saída: nome Maria
  # Saída: idade 31
  # Saída: cidade São Paulo
```

**Modularização**


Esses subprogramas são definidos como procedimentos ou funções. O subprograma é um procedimento quando  executa um conjunto de instruções sem a necessidade de retornar valores pelo seu identificador e é considerado uma função quando o identificador retornar valores.



```python
# prompt: Escreva uma modelo de programa utilizando procedimentos

def saudacao(nome):
  """
  Esta função recebe um nome e imprime uma saudação.
  """
  print("Olá,", nome + "!")

def calcular_soma(a, b):
  """
  Esta função recebe dois números e retorna a soma deles.
  """
  soma = a + b
  print("A soma de", a, "e", b, "é:", soma)

# Chamando os procedimentos
saudacao("Maria")
calcular_soma(5, 3)
```

```python
# prompt: Escreva um modelo de programa utilizando funções

def calcular_area_retangulo(largura, altura):
  """
  Esta função calcula a área de um retângulo.

  Args:
    largura: A largura do retângulo.
    altura: A altura do retângulo.

  Returns:
    A área do retângulo.
  """
  area = largura * altura
  return area

def calcular_perimetro_retangulo(largura, altura):
  """
  Esta função calcula o perímetro de um retângulo.

  Args:
    largura: A largura do retângulo.
    altura: A altura do retângulo.

  Returns:
    O perímetro do retângulo.
  """
  perimetro = 2 * (largura + altura)
  return perimetro

# Obtendo a largura e altura do usuário
largura = float(input("Digite a largura do retângulo: "))
altura = float(input("Digite a altura do retângulo: "))

# Calculando a área e o perímetro
area = calcular_area_retangulo(largura, altura)
perimetro = calcular_perimetro_retangulo(largura, altura)

# Imprimindo os resultados
print("A área do retângulo é:", area)
print("O perímetro do retângulo é:", perimetro)
```

```python
# prompt: Reescreva o anterior utilizando classes

class Retangulo:
  def __init__(self, largura, altura):
    self.largura = largura
    self.altura = altura

  def calcular_area(self):
    """
    Calcula a área do retângulo.

    Returns:
      A área do retângulo.
    """
    return self.largura * self.altura

  def calcular_perimetro(self):
    """
    Calcula o perímetro do retângulo.

    Returns:
      O perímetro do retângulo.
    """
    return 2 * (self.largura + self.altura)

# Obtendo a largura e altura do usuário
largura = float(input("Digite a largura do retângulo: "))
altura = float(input("Digite a altura do retângulo: "))

# Criando um objeto Retangulo
retangulo = Retangulo(largura, altura)

# Calculando a área e o perímetro
area = retangulo.calcular_area()
perimetro = retangulo.calcular_perimetro()

# Imprimindo os resultados
print("A área do retângulo é:", area)
print("O perímetro do retângulo é:", perimetro)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import random

# Texto de exemplo
text = "Python é uma linguagem poderosa. Python é ótimo para análise de dados. Python, Python, Python!"

# Gera a word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='viridis',  # Paleta de cores
    random_state=42      # Para reprodutibilidade
).generate(text)

# Exibe a imagem gerada
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
