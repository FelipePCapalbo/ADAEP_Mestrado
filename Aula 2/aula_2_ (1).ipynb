##  O que são dados?
Dados são valores atribuidos a características de Entidades.

**Entidades Físicas**

Definição: Entidades **físicas são objetos ou coisas que existem no mundo material** e podem ser percebidos pelos nossos sentidos. Elas têm uma presença tangível, ocupam espaço e podem ser manipuladas diretamente.

**Entidades Abstratas**

Definição: Entidades abstratas são conceitos ou ideias que não têm uma existência física tangível. Elas existem no plano das ideias, do pensamento, ou como representações simbólicas, e não podem ser percebidas diretamente pelos sentidos.


## Tipos de dados



###Dados Qualitativos:  
tudo o que se refere à qualidade de algo.

 **Descritivos**: Eles fornecem detalhes sobre as qualidades ou atributos de algo, como cor, textura, comportamento, opinião, etc.

**Categóricos:** São geralmente classificados em categorias.

**Subjetivos:** Muitas vezes dependem da interpretação individual e podem variar de pessoa para pessoa.

**Nominais** → não possuem ordem natural, apenas diferenciam categorias.

**Ordinais** → têm uma ordem lógica, mas sem intervalo numérico definido.

image.png
###Dados Quantitativos:
dados que são mensuráveis ou se referem a números.

**Numéricos:** Podem ser representados em números, facilitando o cálculo e a análise.

**Mensuráveis:** Permitem medir características ou quantidades, como altura, peso, temperatura, idade, etc.
Discretos ou Contínuos:
Discretos: Contam coisas e só assumem valores inteiros (ex: número de pessoas em uma sala).

**Contínuos:** Medem coisas e podem assumir qualquer valor dentro de um intervalo (ex: altura, peso).

**Objetivos:** São baseados em fatos e não estão sujeitos a interpretação pessoal.

image.png





##Dados Estruturados

__Dados estruturados__ são aqueles que seguem um formato predefinido e organizado, facilitando sua captura, armazenamento, gerenciamento e análise.

__Formato predefinido:__ Os dados se encaixam em campos específicos, como nome, idade, endereço, etc.

__Organização:__ Os dados são armazenados em uma estrutura lógica, como tabelas, bancos de dados relacionais ou hierárquicos.

Facilidade de análise: Devido à sua organização, os dados estruturados são facilmente processados por softwares e algoritmos, permitindo a realização de análises estatísticas e a extração de insights.

image.png

##Dados Não Estruturados

__Dados não estruturados__ são aqueles que não seguem um formato predefinido e organizado, ao contrário dos dados estruturados. Eles são mais flexíveis e podem assumir diversas formas, tornando-os mais complexos de capturar, armazenar e analisar.

__Formato variável:__ Não se encaixam em campos específicos e podem ter diferentes tamanhos e estruturas.

__Dificuldade de organização:__ A falta de um formato padrão dificulta a organização e a busca por informações específicas.

__Grande volume:__ Representam a maior parte dos dados gerados atualmente, especialmente com o crescimento do uso de mídias sociais, e-mails e outros canais digitais.

##Dados semi-estruturados

Os __dados semi-estruturados__ representam um ponto intermediário entre os dados estruturados e não estruturados, combinando elementos de ambos. Eles possuem uma certa organização, mas não seguem um esquema rígido como os dados estruturados, e não são tão desordenados quanto os dados não estruturados.

__Organização parcial:__ Possuem uma estrutura interna, como tags ou hierarquias, que permitem identificar e agrupar informações.

__Flexibilidade:__ Permitem a adição de novos elementos ou a modificação da estrutura existente, oferecendo maior flexibilidade do que os dados estruturados.

__Heterogeneidade:__ Podem conter diferentes tipos de dados em um mesmo conjunto, como texto, números e datas.

```python
import requests

# URL da API
url = "https://api.portaldatransparencia.gov.br/api-de-dados/servidores"

# Cabeçalhos
headers = {
    "accept": "*/*",
    "chave-api-dados": "dcaf878d9888c8a2c98a050a50d420e2"
}

pagina = 1
todos_dados = []

while pagina < 50:
    # Parâmetros da requisição
    params = {
        "orgaoServidorExercicio": "26280",
        "anoExercicio": "2025",
        "pagina": pagina
    }

    # Requisição GET
    response = requests.get(url, headers=headers, params=params)

    if response.status_code != 200:
        print(f"Erro {response.status_code}: {response.text}")
        break

    dados = response.json()

    # Se não houver dados, encerra o loop
    if not dados:
        break

    # Concatena os dados recebidos
    todos_dados.extend(dados)

    print(f"Página {pagina} processada, total acumulado: {len(todos_dados)} registros")
    pagina += 1
print(todos_dados)
```

```python
import pandas as pd

# Transforma a variável 'dados' em um DataFrame
df_dados = pd.DataFrame(todos_dados)

# Expande a coluna 'fichasCargoEfetivo' para ter uma linha por ficha
df_explodido = df_dados.explode('fichasCargoEfetivo')

# Extrai 'uorgLotacao' do dicionário aninhado e obtém valores únicos
departamentos_unicos_pandas = df_explodido['fichasCargoEfetivo'].apply(lambda x: x.get('uorgLotacao')).dropna().unique().tolist()

# Itera por cada departamento único
for departamento in departamentos_unicos_pandas:
    print(f"Departamento: {departamento}")
    # Filtra o DataFrame para professores no departamento atual
    professores_no_departamento = df_dados[df_dados['fichasCargoEfetivo'].apply(lambda x: any(ficha.get('uorgLotacao') == departamento for ficha in x))]

    # Extrai e imprime os nomes dos professores neste departamento
    for index, linha in professores_no_departamento.iterrows():
        nome_servidor = linha['servidor']['pessoa']['nome']
        print(f"- {nome_servidor}")
    print("-" * 20) # Separador para clareza
```

```python
import requests
from bs4 import BeautifulSoup

# Fazendo a requisição
url = "https://dep.ufscar.br/index.php/dep-docentes/"
response = requests.get(url)


# Analisando o HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Encontrando todos os elementos com a classe "nome-docente" (ajuste conforme a estrutura da página)
nomes = soup.find_all('h5', class_='person_name')



# Extraindo e imprimindo os nomes
for nome in nomes:
    herf = nome.find('a')
    print(herf.text, herf.get('href'))
```
